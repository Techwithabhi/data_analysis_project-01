# Upload the raw CSV file for Executing EDA portion

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../data/processed/cleaned_data.csv")
df.head()
```

### **Age Segmentation**

```python
def age_group(age):
    if age < 30:
        return "Young"
    elif age < 45:
        return "Mid"
    elif age < 60:
        return "Senior"
    else:
        return "Elder"

df.insert(
    loc=6,                     # index position (0-based)
    column='AgeGroup',
    value=df['Age'].apply(age_group)
)
```

```python
df.head(10)
```

### **Credit Score Bands**

```python
def credit_band(score):
    if score < 500:
        return "Low"
    elif score < 700:
        return "Medium"
    else:
        return "High"

df.insert(
    loc=3,                     # index position (0-based)
    column='CreditBand',
    value=df['CreditScore'].apply(credit_band)
)
```

```python
df.head(10)
```

### **Balance Segmentation**

```python
def balance_group(balance):
    if balance == 0:
        return "Zero"
    elif balance < 100000:
        return "Low"
    else:
        return "High"


df.insert(
    loc=10,                     # index position (0-based)
    column='BalanceGroup',
    value=df['Balance'].apply(balance_group)
)
```

```python
df.head(10)
```

### **Tenure Segmentation**

```python
def tenure_group(t):
    if t <= 3:
        return "New"
    elif t <= 7:
        return "Mid"
    else:
        return "Long"

df.insert(
    loc=9,                     # index position (0-based)
    column='TenureGroup',
    value=df['Tenure'].apply(tenure_group)
)
```

```python
df.head(10)
```

```python
(df.groupby('TenureGroup')['Exited'].mean() * 100).round(2)
```

```python
# Final Check
df[['AgeGroup', 'CreditBand', 'BalanceGroup', 'TenureGroup']].head()
```

## Exporting the EDA CSV data (Final)

```python
df.to_csv("../data/processed/final_data.csv", index=False)
```

## **KPI Analysis**

```python
# Import final data
df = pd.read_csv("../data/processed/final_data.csv")
```

```python
# Overall Churn Rate
df['Exited'].mean() * 100
```

```python
# Churn by Geography
df.groupby('Geography')['Exited'].mean() * 100
```

### Geography Insight

- Germany has the highest churn rate at ~32%, which is nearly double compared to France and Spain (~16%).
- This indicates a major retention issue in the German customer segment.
- France and Spain show relatively stable and lower churn behavior.

```python
# Age Group Analysis
(df.groupby('AgeGroup')['Exited'].mean() * 100).round(2)
```

### Age Group Insight

- Senior customers have the highest churn rate at ~49%, indicating a critical retention issue.
- Elder customers also show elevated churn (~28%), reinforcing risk in older segments.
- Young customers have the lowest churn (~7.5%), making them the most stable segment.
- Churn increases significantly with age, suggesting age-driven behavioral or service gaps.

```python
# Engagement Analysis
(df.groupby('IsActiveMember')['Exited'].mean() * 100).round(2)
```

### Engagement Insight

- Inactive customers have a significantly higher churn rate (~27%) compared to active customers (~14%).
- Inactive users are nearly 2x more likely to churn.
- Customer engagement is a strong indicator of retention.

## **FINAL KPI**

```python
# Are valuable customers leaving?
high_value = df[df['Balance'] > 100000]
(high_value['Exited'].mean() * 100).round(2)
```

## Final Business Insights

- Overall churn rate is ~20%, indicating a significant customer retention challenge.
- Germany shows the highest churn (~32%), making it the most critical geographic risk area.
- Senior customers have an extremely high churn rate (~49%), highlighting a major issue in older customer segments.
- Inactive customers are nearly 2x more likely to churn compared to active users, proving engagement is a key retention factor.
- High-value customers show a churn rate of ~25%, which is a serious business concern as it directly impacts revenue.

## **Data Visualization**

```python
# Churn by Geography
geo_churn = df.groupby('Geography')['Exited'].mean().reset_index()

sns.barplot(data=geo_churn, x='Geography', y='Exited')
plt.title("Churn Rate by Geography")
plt.show()
```

### Churn by Geography (Visualization Insight)

- Germany clearly shows the highest churn rate among all regions.
- France and Spain have significantly lower and similar churn levels.
- This visual confirms that Germany is the primary geographic risk area for customer retention.

```python
# Create Age Group Chart
age_churn = df.groupby('AgeGroup')['Exited'].mean().reset_index()

sns.barplot(data=age_churn, x='AgeGroup', y='Exited')
plt.title("Churn Rate by Age Group")
plt.show()
```

### Churn by Age Group (Visualization Insight)

- Churn increases significantly with age, with Senior customers showing the highest churn rate.
- Nearly half of the Senior segment is leaving, indicating a critical retention issue.
- Younger customers (Young & Mid) demonstrate relatively low churn, making them more stable.
- This trend suggests that older customers may be facing usability, service, or satisfaction challenges.

```python
# Create Engagement Chart
active_churn = df.groupby('IsActiveMember')['Exited'].mean().reset_index()

sns.barplot(data=active_churn, x='IsActiveMember', y='Exited')
plt.title("Churn Rate by Activity Status")
plt.show()
```

### Churn by Activity Status (Visualization Insight)

- Inactive customers show a significantly higher churn rate compared to active customers.
- The churn rate for inactive users is nearly double, highlighting a strong correlation between engagement and retention.
- Active customers are more likely to stay, indicating that regular interaction with the bank reduces churn risk.
- This suggests that improving customer engagement can directly reduce churn.

```python
# High-Value Customer Churn
high_value = df[df['Balance'] > 100000]

sns.barplot(x=['High Value'], y=[high_value['Exited'].mean()])
plt.title("Churn Rate of High-Value Customers")
plt.show()
```

### High-Value Customer Churn (Visualization Insight)

- High-value customers show a churn rate of approximately 25%, which is a significant business concern.
- These customers contribute heavily to revenue, so their loss has a direct financial impact.
- Retaining high-value customers should be a priority through personalized services and targeted retention strategies.
